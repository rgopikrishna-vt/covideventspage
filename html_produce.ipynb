```python
covid_apple_text_dict = {}
covid_apple_head_dict = {}
appledf = df_covid[df_covid.ENTITY_NAME.str.contains('Apple')]
events_covid_apple = df_covid[df_covid.ENTITY_NAME.str.contains('Apple')].TYPE.unique()
for event in events_covid_apple:
    covid_apple_text_dict[event] = appledf[appledf.TYPE==event].EVENT_TEXT.unique()
    covid_apple_head_dict[event] = appledf[appledf.TYPE==event].HEADLINE.unique()
```
